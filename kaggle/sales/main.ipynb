```python
import os
import datetime

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Forecast:
    def __init__(self):
        cwd = os.getcwd()
        for dataset in ['holidays_events','oil','sample_submission','stores','test','train','transactions']:
            df = pd.read_csv(os.path.join(cwd, 'data', dataset + '.csv')).rename({'date':'ds'}, axis=1)
            try:
                df['ds'] = pd.to_datetime(df['ds'])
            except:
                pass
            setattr(self, dataset, df)
        
```

```python
fc = Forecast()
```

```python
fc.train.merge(fc.transactions,how='left',on=['ds','store_nbr'])
```
